```python
import pickle
import os

def rePickle(name):
    pickle3 = pickle.load( open( "weaksymBidentatesProj/" + str(name), "rb" ) )
    pickle.dump(pickle3, file=open("weaksymBidentatesProj/" + str(name) + '2', "wb"), protocol=2)
    
for file in os.listdir("weaksymBidentatesProj/"):
    print(file)
    rePickle(file)
```
